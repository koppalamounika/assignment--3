3.Compute outer product of 2 vectors. Eg:

original matrix: [[1, 0], [0, 1]] [[1, 2], [3, 4]] Outer product of the said two vectors: [[1 2 3 4] [0 0 0 0] [0 0 0 0] [1 2 3 4]]


```
import numpy as np
p=[[1,0],[0,1]]
q=[[1,2],[3,4]]
print("ORIGINAL MATRIX IS:")
print(p)
print(q)
print("THE OUTER PRODUCT OF GIVEN VECTORS IS :\n",np.outer(p,q))
```

4.Write a NumPy program to compute the condition number of a given matrix. Original matrix: [[1 2] [3 4]] Condition number of the said matrix: 14.9330343737

```
import numpy as np
m = np.array([[1,2],[3,4]])
print("Original matrix:")
print(m)
result =  np.linalg.cond(m)
print("Condition number of the said matrix:")
print(result)
```
